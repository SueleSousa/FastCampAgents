```python
import warnings

# Ignora avisos
warnings.filterwarnings('ignore')

# Importa as classes Agent, Task e Crew da biblioteca crewai
from crewai import Agent, Task, Crew

import os
from utils import get_openai_api_key

# Obtém a chave da API do OpenAI usando uma função utilitária
openai_api_key = get_openai_api_key()
# Define o nome do modelo OpenAI a ser usado
os.environ["OPENAI_MODEL_NAME"] = 'gpt-3.5-turbo'

# Cria um agente chamado 'planner' que atua como uma agência de viagens
planner = Agent(
    role="Travelling Agency",
    goal="Give a overview based on expendings to {topic}",
    backstory="You're working on a travel agency and are doing a research "
              "about the place: {topic}. You collect price information about the plane ticket, accommodations, also necessities "
              "and the local weather, so people can plan a good vacation",
    allow_delegation=False,  # Não permite delegação de tarefas
    verbose=True  # Ativa a saída detalhada
)

# Cria um agente chamado 'guide' que atua como um guia local
guide = Agent(
    role="You are a local guide",
    goal="To guide the traveller at the local landmarks",
    backstory="You're working on a travelling agency as a tourist guide. "
              "You're giving information about the landmarks at {topic}. "
              "You make a list of the most importants landmarks. "
              "Your input is give the average time spending on each landmark, which is next to each. "
              "You also provides a list of typical food. "
              "You also give information about the transport to each landmark. "
              "Also give the idea of how much expend in each landmark.",
    allow_delegation=False,  # Não permite delegação de tarefas
    verbose=True  # Ativa a saída detalhada
)

# Define uma tarefa para o agente 'planner'
plan = Task(
    description=(
        "1. Give information about plane tickets price to {topic}\n"
        "2. Give information about accommodations price\n"
        "3. Inform about weather e better months to visit"
    ),
    expected_output="A average total cost of a travel for the chosen place "
                    "with the information about weather, vaccines, and other must have to ",
    agent=planner,  # A tarefa é atribuída ao agente 'planner'
)

# Define uma tarefa para o agente 'guide'
guide_list = Task(
    description=(
        "1. You're giving information about the top 7 landmarks at {topic}.\n"
        "2. You give the average time spending on each landmark, which is next to each other.\n"
        "3. You also provide a list of typical food.\n"
        "4. You also give information about the transport to each landmark.\n"
        "5. Also give the idea of how much expend in each landmark."
    ),
    expected_output="A list very complete to a traveller with landmarks, "
                    "prices, transportation and expenses at which local",
    agent=guide,  # A tarefa é atribuída ao agente 'guide'
)

# Cria uma equipe com os agentes e as tarefas definidas
crew = Crew(
    agents=[planner, guide],  # Lista de agentes
    tasks=[plan, guide_list],  # Lista de tarefas
    verbose=2  # Nível de detalhamento da saída
)

# Inicia a execução
```

```python
# Inicia a execução da equipe com o tópico "Rio de Janeiro"
result = crew.kickoff(inputs={"topic": "Rio de Janeiro"})
```
